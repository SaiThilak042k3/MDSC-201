##Walmart Dataset


```r
#Importing the dataset
df = read.csv("/content/drive/MyDrive/RM-LAB-2024/Walmart.csv")

#Dropping the date feature because it doesn't affect weekly sales

df = subset(df, select = -c(Date))
```

```r
#Setting the seed for randomness
set.seed(123)

#Setting the split ratio of training set and test set and also the target variable
split = sample.split(df$Weekly_Sales, SplitRatio = 0.8)

#printing the split
print(split)
```

```r
#Splitting the dataset df into Training and Testing set
training_set = subset(df, split== TRUE) #training set
test_set = subset(df, split == FALSE) #test set
```

```r
#Fitting the Multiple Linear Regression Model using Training Set
regressor = lm(formula = Weekly_Sales ~ ., data = training_set)

#printing the regressor model
print(regressor)
```

```r
#Predicting the Test Set Results
y_pred = predict(regressor, newdata = test_set)

#printing the predicted vaules
print(y_pred)
```

```r
#checking the Accuary of the Model
y_actual = test_set$Weekly_Sales

#printing the accuary
print(y_actual)
```

##Own Data Set

**FISHER DATASET**

This dataset is a record of 7 common different fish species in fish market sales. With this dataset, a predictive model can be performed using machine friendly data and estimate the weight of fish can be predicted.

source - https://www.kaggle.com/datasets/tarunkumar1912/fish-dataset1/data

```r
#Read the dataset using read.csv
df <- read.csv("/content/drive/MyDrive/RM-LAB-2024/Fish_dataset.csv")

#Display the first few rows of the dataset
head(df)
```

```r
#Setting the seed for randomness
set.seed(123)

#Setting the split ratio of training set and test set and also the target variable
split = sample.split(df$Weight, SplitRatio = 0.8)

#printing the spilted data
print(split)
```

```r
#Splitting the data set into Training and Testing set
training_set = subset(df, split== TRUE) #training data
test_set = subset(df, split == FALSE) #testing data

#Fitting the Multiple Linear Regression Model using Training Set
regressor = lm(formula = Weight ~ ., data = training_set)
print(regressor)
```

```r
# Predicting the Test Set Results
y_pred = predict(regressor, newdata = test_set)
print(y_pred)
```

```r
#Calculating the Accuracy of the model
y_actual = test_set$Weight
print(y_actual)
```
